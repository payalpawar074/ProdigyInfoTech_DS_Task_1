# Task 1

## Import required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv(
    "Population_task_1.csv",
    skiprows=4
)
```

## View first 5 rows

```python
df.head()
```

## Check Column

```python
df.columns
```

## Drop Unwanted Column

```python
df = df.drop(columns=['Unnamed: 69'])
```

## Verify columns again

```python
df.columns
```

## Select Year for Analysis

```python
year = '1975'
```

## Convert Data Type

```python
df[year] = pd.to_numeric(df[year], errors='coerce')
```

## Handle Missing Values

```python
year_data = df[['Country Name', year]].dropna()
```

## Create Histogram (Continuous Variable)

```python
# List of countries for the graph
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Convert 1975 column to numeric
df['1975'] = pd.to_numeric(df['1975'], errors='coerce')

# Filter the DataFrame for selected countries
filtered_data = df[df['Country Name'].isin(countries_to_plot)]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(filtered_data['Country Name'], filtered_data['1975'])
plt.xlabel('Country Name')
plt.ylabel('Population in 1975')
plt.title('Population of Selected Countries in 1975')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Create Bar Chart (Categorical Variable)

```python
top_10 = year_data.sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10['Country Name'], top_10[year])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populated Countries in 1975')
plt.xticks(rotation=45)
plt.show()
```
